```python
df = pd.read_excel('/content/dummy.xlsx')
df.head()
```

```python
df.dtypes
```

```python
df = pd.get_dummies(df, columns=['Reliabilitas', 'Asuransi', 'Aset', 'Empati', 'Responsivitas'])
df.head()
```

```python
column_list = [col for col in df.columns if col != 'content']
column_list
```

```python
texts = df['content'].tolist()
# texts


aspek = df[column_list].values.tolist()
labels = np.array(aspek)
```

Pertanyaan:
1. Apakah Label perlu untuk di one-hot-encoding?


```python
# inisialisasi objek stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()
clean_data = []

for text in texts:
    # lemmatization
    teks_lemmatized = stemmer.stem(text)
    clean_data.append(teks_lemmatized)
print(clean_data)
```

```python
# Split data menjadi data latih dan data uji
train_texts, test_texts, train_labels, test_labels = train_test_split(clean_data, labels, 
                                                                      test_size=0.2, 
                                                                      random_state=42)
```

```python
len(train_texts)
```

```python
len(test_texts)
```

```python
# Tokenisasi dan padding teks
oov_tok ="<OOV>"
tokenizer = Tokenizer(num_words=10000, oov_token=oov_tok)
tokenizer.fit_on_texts(train_texts)
train_sequences = tokenizer.texts_to_sequences(train_texts)
test_sequences = tokenizer.texts_to_sequences(test_texts)
train_padded_sequences = pad_sequences(train_sequences, maxlen=50)
test_padded_sequences = pad_sequences(test_sequences, maxlen=50)
```

```python
train_padded_sequences
```

```python
# Membuat model
input_size = 50
hidden_layer_size = 32
output_size = len(column_list)
lstm_dim = 32

model = tf.keras.models.Sequential([
    tf.keras.layers.Embedding(10000, 16, input_length=input_size),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(lstm_dim)),
    # tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(hidden_layer_size, activation='relu'),
    tf.keras.layers.Dense(output_size, activation='sigmoid')
])
```

```python
# Compile model
from tensorflow import keras
from tensorflow.keras import layers

model.compile(optimizer= keras.optimizers.Adam(learning_rate=0.01),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Training model

model.fit(train_padded_sequences, train_labels, epochs=100, batch_size=1, validation_data=(test_padded_sequences, test_labels))
```

```python


# Prediksi pada teks baru
new_texts = ["Sudah bayar tapi masih ditagih debt colector! Aneh! Menyesal pinjam dan tidak akan pinjam lagi!!",
             "Aduuhh payah ke bank aja 150jt ke aplikasi ini gak di ACC",
             "Aplikasi gak jelas belum pernah pake udah error aja"]

new_sequences = tokenizer.texts_to_sequences(new_texts)
new_padded_sequences = pad_sequences(new_sequences, maxlen=50)
new_predictions = model.predict(new_padded_sequences)

print(new_predictions)
```
